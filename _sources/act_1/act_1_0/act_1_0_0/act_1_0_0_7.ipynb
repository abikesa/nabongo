Instagram quantifies "engagement" through various metrics that measure user interaction with content. These metrics include but are not limited to:

1. **Likes**: The number of likes a post receives.
2. **Comments**: The number of comments on a post.
3. **Shares**: The number of times a post is shared, either through direct messages or by posting it to one's own story.
4. **Saves**: The number of times a post is saved for later viewing.
5. **Story Views**: The number of views a Story receives.
6. **Click-through Rate (CTR)**: For advertisements or links in the bio, how often users click on them.
7. **Follows**: The number of new followers gained as a result of a post or a series of posts.
8. **Impressions**: The total number of times a post or story has been seen.
9. **Reach**: The number of unique accounts that have seen a post or story.
10. **Engagement Rate**: Often calculated as a percentage, it's the total engagement (likes + comments + shares + saves) divided by the total followers, multiplied by 100. This helps normalize the engagement based on the size of the follower base.

Business accounts and influencers often use these metrics to evaluate the success of their posts and overall strategy. These engagement metrics help both Instagram and the account owner understand how well the content is resonating with the audience.
