## Abstract

We aim to develop an Artificial Intelligence (AI)-based risk calculator that predicts perioperative and long-term risks in older adults (aged ≥ 55) who are potential live kidney donors. This tool will assist healthcare providers in making data-driven decisions during donor selection and will provide evidence-based insights for donor-provider discussions. Since we already have a supervised model of a similar risk calculator, we will use this as a starting point and augment it with additional data. We will then validate the model using existing datasets and prospective data. Finally, we will develop a user-friendly interface for healthcare providers. Upon successful development, the tool will be made available to healthcare institutions as a Software as a Service (SaaS) model.
