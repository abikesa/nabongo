+ these are the tokens we're talkin' 'bout with the chatbot
+ please understand that tokens are the words that make up the language
+ but more specifically, their meaning is defined by the context in which they are used
+ so, for example, the word "run" can mean a lot of things
+ it can mean to move quickly on foot (as in "run, forest, run!") 
+ or it can mean to operate a machine (as in "run the dishwasher")
+ perhaps you mean to manage a business (as in "run a company")?
+ we are talking probabilistically here, after some opening context
   - the word "run" is more likely to mean "operate a machine" than "move quickly on foot" 
   - in the context of a conversation about dishwashers
+ as such, the meaning of a word is defined by the context in which it is used
+ context is defined by the tokens that come before it (and after it)
+ when we tokenize a sentence, we are breaking it down into its component parts
+ maybe we might tokenize the alphabet into letters (a, b, c, ..., z)
+ or maybe we might tokenize a sentence into words (the, quick, brown, fox, jumps, over, the, lazy, dog)
+ these tokens are the building blocks of language and assume meaning in context (i.e., positions relative to other tokens)
